````python
from pandas_datareader import data, wb
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters() # Allow matplotlib have access to timestamp 
import matplotlib.ticker as mplticker
import matplotlib.dates as mdates
````

```pandas_datareader``` has a submodule for world bank ```.wb```, and the country code is provided in the list ```pdr.wb.country_codes```.

````python
# use wb submodule, which means worldbank
start = 2000
end = 2019
dat = wb.download(indicator='NE.DAB.TOTL.ZS',  # Gross National Expenditure (% Of GDP) 
                  country = pdr.wb.country_codes[-1],  # we want all country's data
                  start = start, end = end)
````

````python
dat.columns = ['GNE/GDP %']
````

````python
from matplotlib.dates import YearLocator
````

````python
fig, ax = plt.subplots(figsize = (14, 7))
natList = ['China', 'United States', 'United Kingdom', 
           'Finland', 'Germany', 'Poland', 'Japan', 'Hong Kong SAR, China',
           'Singapore', 'Iceland']
for i in range(len(natList)):
    ax.plot(dat.loc[natList[i]][::-1], label = natList[i])
ax.legend()
ax.xaxis.set_tick_params(rotation=45)
````

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import pandas_datareader as pdr
```

# <font face="gotham" color="purple"> Demographics

```python
# Load data
start = dt.datetime(1951, 1, 1)
end = dt.datetime.today()
df = pdr.data.DataReader(['SPDYNTFRTINPOL', 'SPADOTFRTPOL'], 'fred', start, end) 
df.columns = ['TFR','Adolescent FR']
```

Total fertility rate is the number of children a woman will give birth to during her whole childbearing years. 

Adolescent fertility rate is the number of childbirth of 100 teenagers age between 15-19.

```python
fig, ax = plt.subplots(figsize = (13, 8))
ax.plot(df['TFR'], color = 'CornflowerBlue', alpha = 1, label = 'Total Fertility Rate')
ax.plot(df['Adolescent FR']*.1, color = 'DarkGoldenRod', alpha = .7,label = 'Adolescent Fertility Rate')

ax.legend(fontsize = 13)
ax.set_title('TFR and AFR in Poland', size = 15)
ax.grid()
plt.show()
```

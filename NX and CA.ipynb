```python
from pandas_datareader import data, wb
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters() # Allow matplotlib have access to timestamp 
import matplotlib.ticker as mplticker
```

```python
start = dt.datetime(1975, 1, 1)
end = dt.datetime.today()
columns = ['Gross Savings', 'Gross Domestic Investment','CA Imbalance','Net Export','GDP']
df = pdr.DataReader(['GSAVE','W170RC1Q027SBEA','NETFI','NETEXP','GDP'], 'fred', start, end)
df.columns = columns
```

```python
NetSaving = df['Gross Savings']- df['Gross Domestic Investment']
```

```python
fig, ax = plt.subplots(figsize = (12, 7))
ax.plot(NetSaving, label = 'Net Saving')
ax.plot(df['Net Export'], label = 'NX')
ax.plot(df['CA Imbalance'], label = 'CA')
ax.legend()
plt.show()
```
